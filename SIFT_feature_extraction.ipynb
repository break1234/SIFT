```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Load the image

```python
img = cv2.imread('images/1.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #Correct BGR to RGB channel
plt.imshow(img)
```

Convert image to Graysclae

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

Instantiate SIFT feature extractor and find keypoints

```python
sift = cv2.xfeatures2d.SIFT_create()
(kps, descs) = sift.detectAndCompute(gray,None)
```

Draw keypoints on the image

```python
cv2.drawKeypoints(gray,kps,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.imwrite('images/sift_keypoints.png',img)
plt.imshow(img)
```
